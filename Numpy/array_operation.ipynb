Broadcasting


```python
l=[10,20,30,40,50]
arr=np.array(l)
arr
```

```python
arr+10
```

```python
arr2=np.arange(1,26).reshape(5,5)
arr3= np.arange(1,26).reshape(5,5)
arr2
```

```python
arr2*arr3
```
