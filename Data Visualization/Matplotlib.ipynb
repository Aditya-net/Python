Matplotlib basics

```python
x= np.linspace(0,5,11)
x
y = x**2
y
```

```python
plt.plot(x, y)
plt.title("HELLO")
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

Subplots in matplotib

```python
plt.subplot(2,2,1)
plt.plot(x,x+2)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x,x)
```

**Object Oriented way**

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.

figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch (pixel per inch).

```python
fig = plt.figure(figsize=(8,4),dpi=900)
axis1=fig.add_axes([0.1,0.1,0.8,0.8])# left, bottom, width, height
axis1.plot(x,y)
axis2=fig.add_axes([0.3,0.4,0.3,0.3])
axis2.plot(y,x)
```
